# Bayesian Linear Regression

```julia
using DataFrames
using Distributions
using LinearAlgebra
using Measures
using Plots
using PlotThemes
using Random
Random.seed!(111)

pyplot()
theme(:orange)
```

## Simulate Dataset

```julia
function simulate(n::Int64 = 1000, w0::Float64 = -.3, w1::Float64 = -.6, alpha::Float64 = 1 / 5)
    x = rand(Uniform(-1, 1), n)

    w = [w0; w1]
    A = [ones(size(x)) x]
    
    y = A * w + rand(Normal(0, alpha), size(x))
    return (x, y, A, alpha)
end

function posterior(y::Array{Float64, 1}, A::Array{Float64, 2}, alpha::Float64, beta::Float64 = (1 / 2)^2)
    sigmainv = alpha * A'A + beta * I;
    sigma = sigmainv^(-1);
    mu = alpha * sigma * A' * y;
    return (mu, sigma)
end
```

```julia
x, y, A, a = simulate();
mu, sigma = posterior(y, A, a);
```

```julia
plt = plot(
    x, y, 
    xlab = "x", 
    ylab = "y", 
    title = "Fitted lines with weights ~ p(weights|data)",
    markershape = :circle,
    markersize = 5,
    markeralpha = 0.5,
    markerstrokewidth = 0,
    linealpha = 0, 
    legend = false,
    dpi = 200, 
    size = (700, 432),
    margin = 10mm
)

plot!(annotate = (0.5, 0.5, text("hey")))
```

```julia
plt = plot(
    x, y, 
    xlab = "x", 
    ylab = "y", 
    title = "Fitted lines with weights ~ p(weights|data)",
    markershape = :circle,
    markersize = 5,
    markeralpha = 0.5,
    markerstrokewidth = 0,
    linealpha = 0, 
    legend = false,
    dpi = 200, 
    size = (700, 432),
    margin = 10mm
)

Random.seed!(111)
anim = @animate for i = 1:110
    w = rand(MvNormal(mu, sigma))
    
    if i < 100
        plot!(plt, x, A * w, linewidth = 1.5, linealpha = 0.7, linecolor = colorant"#84ae5a")
    else
        plot!(plt, x, A * mu, linewidth = 1.5, linealpha = 0.1, linecolor = colorant"#dc3835")
    end
end

gif(anim, "2018-10-10-p1.gif", fps = 15)
```
